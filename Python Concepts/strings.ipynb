## Creating a string 

```python
# single quotes & double quote 
# Mostly Used based on preference or need.
single_quoted = 'This is a string.'
double_quoted = "This is also a string."
print(single_quoted,end='\n\n')
print(double_quoted)
```

```python
# \ 
print("This is a path: C:\\Users\\Username\\Documents")

# \n - new line 
print('frist line \nsecond line')

# \t - tab space
print("Name:\tJohn")

# \b - backspace character 
print("Back \bspace")
```

## String Slicing
- access a range of characters from a string using the slicing operator [:]
- String slicing allows you to extract a portion of a string.
- The syntax is string [start:stop:step].
- Negative indices count from the end of the string.

```python
# Example string
string_example = "python"

print("string_example[0]: " + string_example[0])

print("string_example[2:]: " + string_example[2:])

print("string_example[:3]: " + string_example[:3])

print("string_example[1:4]: " + string_example[1:4])

print("string_example[0:4:2]: " + string_example[0:4:2])

print("string_example[-2:]: " + string_example[-2:])

print("string_example[:-3]: " + string_example[:-3])

print("string_example[-4:-1]: " + string_example[-4:-1])

print("string_example[2::-1]: " + string_example[2::-1])

print("string_example[:-4:-1]: " + string_example[:-4:-1])

print("string_example[::-1]: " + string_example[::-1])
```

## String Formatting

```python
# 1. "old-style" - % operator
name = 'shadow'
print('hello, %s' %name)
```

```python
#2. "new-format" - str.format
name = 'shadow'
print('hello, {}'.format(name))

a = 'see'
b = 'you'
c = 'soon'

print('{0} {1} {2}'.format(a,b,c),end= '\n\n')
print('{1} {0} {2}'.format(a,b,c),end = '\n\n')
print('{2} {1} {0}'.format(a,b,c))
```

```python
# 3. String interpolation/f-string 
name = 'shadow'
print(f'hello ,{name}')

# can do math operation also
a = 10
b = 7
print(f'{a} - {b} = {a - b}')


# also f in the print is not case-sensitive F can also be used 
```

```python
# 4.template 
from string import Template
s = Template('$car is a $type car')
print(s.safe_substitute(car = 'R32',type = 'JDM'))
```

when to use :
Template formatting makes it easier to translate app content into different languages. Translators focus on fixed template structures, like placeholders, without dealing with changing details. This separation simplifies the translation process.

## string methods
- capitalize(): Converts the first character to uppercase.
- lower(): Converts all characters to lowercase.
- upper(): Converts all characters to uppercase.
- count(substring): Counts occurrences of a substring.
- find(substring): Returns the index of the first occurrence of a substring.
- replace(old, new): Replaces occurrences of a substring with another substring.
- split(delimiter): Splits the string into a list of substrings.
- strip(): Removes leading and trailing whitespaces.


 [for-more-methods](https://www.geeksforgeeks.org/python-string-methods/)

```python
# Example string
input_string = "shadow"

# capitalize()
print(f"capitalize(): {input_string.capitalize()}")

# lower()
print(f"lower(): {input_string.lower()}")

# upper()
print(f"upper(): {input_string.upper()}")

# count()
print(f"count('d'): {input_string.count('d')}")

# find()
print(f"find('o'): {input_string.find('o')}")

# replace()
print(f"replace('o', 'x'): {input_string.replace('o', 'x')}")

# split()
print(f"split('d'): {input_string.split('d')}")

# strip()
print(f"strip(): {input_string.strip()}")
```
