# SETS

- almost same as lists and same operations

```python
# set has a forzenset version which is the immutable set
# can be created using fronzeset() constructor

set1 = frozenset([1, 2, 3])
set2 = frozenset([3, 4, 5])

union_set = set1.union(set2)
intersection_set = set1.intersection(set2)
difference_set = set1.difference(set2)
is_disjoint = set1.isdisjoint(set2)
length = len(set1)
element_present = 2 in set1

print("Union:", union_set)
print("Intersection:", intersection_set)
print("Difference:", difference_set)
print("Is Disjoint:", is_disjoint)
print("Length:", length)
print("Element 2 Present:", element_present)
```

#### THESE WORK FOR BOTH SETS AND FROZENSETS
