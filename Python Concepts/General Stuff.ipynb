# Comments

```python
# comments in python 
# note: comments will not be displaced in the output.
# single line comment use (#) 

# single line comment
print("For single line comment just use #")

# multi line comment
"""
this 
is 
a multi-line 
comment """
print("For a multi-line comment, use \"\"\" or '''")

# docs comments
# these comments are used in functions, methods, classes or modules
# these comments are used for documentation purpose not for execution

def greet(name):
    """
    This function greets the person whose name is passed as a parameter.
    
    Parameters:
    name (str): The name of the person to greet.
    
    Returns:
    str: A greeting message.
    """
    return f"Hello, {name}!"

name = 'Shadow'
greet(name)
```
